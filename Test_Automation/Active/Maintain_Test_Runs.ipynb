```python
##########################
#Delete Test Runs and Their Results
##########################

# ------------------------------------------
# Configuration
# ------------------------------------------

RUNS_TABLE = "test_automation_runs"
RESULTS_TABLE = "test_automation_results"

# ------------------------------------------
# Delete Function
# ------------------------------------------

def delete_run_and_results(run_id: str):
    
    print(f"Deleting run: {run_id}")
    
    # Count results before delete (for logging)
    result_count = spark.sql(f"""
        SELECT COUNT(*) AS cnt
        FROM {RESULTS_TABLE}
        WHERE run_id = '{run_id}'
    """).collect()[0]["cnt"]
    
    print(f"Found {result_count} result records to delete")

    # Delete results first (child records)
    spark.sql(f"""
        DELETE FROM {RESULTS_TABLE}
        WHERE run_id = '{run_id}'
    """)
    
    print("Results deleted")

    # Delete run record
    spark.sql(f"""
        DELETE FROM {RUNS_TABLE}
        WHERE run_id = '{run_id}'
    """)

    print("Run deleted")

    print("Delete operation completed")


# ------------------------------------------
# Execute
# ------------------------------------------
run_id_to_delete = "6c51a77a3-0d9b-41e7-93e3-264fa546b342"
delete_run_and_results(run_id_to_delete)
```
